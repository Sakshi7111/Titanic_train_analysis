# COUNTPLOT FOR SURVIVED OR NOT SURVIVED(Q1)

```python
import seaborn as sns
```

```python
sns.countplot("Survived",data=titanic_train_data,palette='nipy_spectral')
```

# COUNTPLOT FOR SURVIVED OR NOT SURVIVED SHOWING SEX USING HUE COMMAND(Q2)

```python
sns.countplot('Survived',data=titanic_train_data,hue='Sex',palette='ocean')
```

# COUNTPLOT FOR SURVIVED OR NOT SURVIVED SHOWING PASSENGER CLASS USING HUE COMMAND (Q3)

```python
sns.countplot('Survived',data=titanic_train_data,hue='Pclass',palette='prism')
```

# COUNTPLOT FOR SIBSP IN A GRAPH REPRESENTATION(Q4)

# BOXPLOT FOR PASSENGER CLASS AND AGE SHOWING X=PCLASS AND Y=AGE ALONG WITH OUTLIARS(Q5)
